```python
# Import library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from scipy import stats
import math
```

```python
# Upload file dataset di Google Colab
from google.colab import files
uploaded = files.upload()
```

```python
# Baca dataset
df = pd.read_csv(list(uploaded.keys())[0])

# Tampilkan 5 data pertama
df.head()
```

```python
# Melihat nama kolom
df.columns
```

```python
# Cek jumlah data berdasarkan Genre
df['genre'].value_counts()
```

```python
# 1. Hitung kemunculan setiap genre
genre_counts = df['genre'].value_counts()

# 2. Ambil genre yang muncul minimal 2 kali (agar bisa dibagi ke train dan test)
valid_genres = genre_counts[genre_counts >= 2].index

# 3. Filter dataframe agar hanya menyisakan genre yang valid
df_filtered = df[df['genre'].isin(valid_genres)]

# 4. Jalankan Stratified Random Sampling menggunakan data yang sudah difilter
train_data, stratified_sample = train_test_split(
    df_filtered,
    test_size=0.2,
    stratify=df_filtered['genre'],
    random_state=31 # Sesuai dengan 3 digit terakhir NIM kamu
)

# Tampilkan hasil sampling
stratified_sample
```

```python
# Ambil hanya 30 data teratas dari hasil sampling
sample_30 = stratified_sample.head(30)

# Tampilkan 30 data
sample_30
```

```python
# Cek kolom Budget
sample_30['budget'].head()
```

```python
# Mengubah kolom Budget menjadi angka
sample_30['budget'] = pd.to_numeric(sample_30['budget'], errors='coerce')

# Menghapus data Budget yang kosong
budget_sample = sample_30['budget'].dropna()

budget_sample
```

```python
# Menghitung jumlah data, rata-rata, dan standar deviasi
n = len(budget_sample)
mean_budget = np.mean(budget_sample)
std_budget = np.std(budget_sample, ddof=1)

print("Jumlah data:", n)
print("Rata-rata Budget:", mean_budget)
print("Standar Deviasi:", std_budget)
```

```python
# Estimasi mean populasi menggunakan T-Distribution
# Confidence level 95%

confidence_level = 0.95
alpha = 1 - confidence_level
dfree = n - 1

t_value = stats.t.ppf(1 - alpha/2, dfree)

margin_error = t_value * (std_budget / math.sqrt(n))

lower_bound = mean_budget - margin_error
upper_bound = mean_budget + margin_error

print("T Value:", t_value)
print("Margin of Error:", margin_error)
print("Estimasi Mean Budget:")
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
# Kesimpulan otomatis

print("Kesimpulan:")
print(f"Berdasarkan 30 data sampel hasil Stratified Random Sampling berdasarkan Genre,")
print(f"dengan random_state = 31 sesuai NIM, diperoleh rata-rata Budget Film sebesar {mean_budget:.2f}.")
print(f"Dengan tingkat kepercayaan 95%, estimasi rata-rata Budget Film populasi berada")
print(f"di antara {lower_bound:.2f} sampai {upper_bound:.2f}.")
```
